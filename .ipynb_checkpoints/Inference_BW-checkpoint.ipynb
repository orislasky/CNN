```python
from tensorflow.keras.preprocessing import image
from tensorflow.keras import backend as K
from tensorflow.keras.preprocessing.image import img_to_array, load_img

x = img_to_array(new_image)
x = np.expand_dims(x, axis=0)
preds = model.predict_classes(x)
prob = model.predict(x)

if 1 in preds:
    print("We are sorry to inform you that you have a tumer with a",prob.max()*100,"percent chance that we are correct") 
else:
    print("YAY you dont have a tumer but you should probably go to a doctor since we are only", prob.max()*100, "percent confident that we are right")
#print(preds, prob.max())
```
